#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
df_cinturon_sur['precios_normalizados'] = 0 

for poblacion in df_cinturon_sur['level5'].unique():

    poblacion_unica = df_cinturon_sur[df_cinturon_sur['level5']==poblacion]
    media_precio = poblacion_unica['price'].mean()
    desviacionestandar_precio = poblacion_unica['price'].std()
    df_cinturon_sur.loc[df_cinturon_sur['level5']==poblacion, 'precios_normalizados'] = (
        poblacion_unica['price'] - media_precio) / desviacionestandar_precio
    
plt.figure(figsize=(12,8))
for poblacion in df_cinturon_sur['level5'].unique():
    poblacion_unica = df_cinturon_sur[df_cinturon_sur['level5']==poblacion]
    plt.hist(poblacion_unica['precios_normalizados'], bins=20, alpha=0.5, label=poblacion, density=True)

plt.title("Histograma de precios normalizados por cada población")
plt.xlabel("Precio normalizado")
plt.ylabel("Densidad")
plt.legend(title="Poblacion")
plt.show()
```

Las conclusiones que podemos sacar con esta gráfica es que Leganés es el municipio más caro, seguido de cerca por Fuenlabrada y alcorcón. En el caso de Getafe, vemos que se encuentra más alejado de estos precios.

```python
from ipyleaflet import Map, CircleMarker, basemaps

df_cinturon_sur['latitude'] = pd.to_numeric(df_cinturon_sur['latitude'].str.replace(',', '.'))
df_cinturon_sur['longitude'] = pd.to_numeric(df_cinturon_sur['longitude'].str.replace(',', '.'))

map_center = [df_cinturon_sur['latitude'].mean(), df_cinturon_sur['longitude'].mean()]
mapa_ubicaciones = Map(center=map_center, zoom=10, basemap=basemaps.OpenStreetMap.Mapnik)

colors = {
    'Fuenlabrada': 'blue',
    'Leganés': 'green',
    'Getafe': 'orange',
    'Alcorcón': 'purple'
}

def add_marcador(row):
    marcador = CircleMarker(
        location=(row['latitude'], row['longitude']),
        radius=5,
        color=colors.get(row['level5']),
        fill=True,
        fill_color=colors.get(row['level5']),
        fill_opacity=0.7
    )
    mapa_ubicaciones.add_layer(marcador)

df_cinturon_sur.apply(add_marcador, axis=1)

mapa_ubicaciones

#NOTA - hay que mover el mapa un poco hacia arriba para ver los puntitos. Si no aparece el mapa por favor recargar y ejecutar.
```
